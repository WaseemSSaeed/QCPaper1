```python
import math
import random
import numpy as np
```

```python
A = 1; K = 10000; m = 1;
OMEGA = math.sqrt(K/m)
Phi = 0
```

```python


import matplotlib.pyplot as plt

xovert = []
iS = []
for i in range(0, 600):
    t = random.uniform(0, 9999999)
    x = A * math.sin(OMEGA * t + Phi)
    xovert.append(x*t)
    iS.append(i)
    plt.plot(i, x*t, color='green', marker='o')
    
plt.show()
```

```python
from openpyxl import Workbook
```

```python
wb = Workbook()
ws = wb.active
ws["A1"] = "shot"
ws["B1"] = "x*t"
for no in range(0, len(iS)):
    #print(str(no))
    ws["A"+str(no+2)] = iS[no]
    ws["B"+str(no+2)] = xovert[no]
wb.save("xTimesTime.xlsx")
```
