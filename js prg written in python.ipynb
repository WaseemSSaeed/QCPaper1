```python
import math, random
import numpy as np
A = 1; K = 10000; m = 1;
OMEGA = math.sqrt(K/m)
Phi = 0
_totalUPS = 0
_totalDowns = 0
__evolve = {"ups": [], "downs": [], "time": []}
__evolveCount = 0
__down0up1 = []

A2 = 1; K2 = 10000; m2 = 1;
OMEGA2 = math.sqrt(K2/m2)
Phi2 = 0
_totalUPS2 = 0
_totalUPS2 = 0
_totalDowns2 = 0
__evolve2 = {"ups": [], "downs": [], "time": []}
__evolveCount2 = 0
__2down0up1 = []
```

```python
_additiveUps = 0
_additiveDowns = 0
shotsCount = 50000
for shot in range(0, shotsCount):
    _ups = 0
    _downs = 0
    t = random.uniform(0, 9999999)
    x = A * math.sin(OMEGA * t + Phi)
    if x > 0:
        _ups += 1
        _totalUPS += 1
        _additiveUps += 1
        __evolve["ups"].append(((_totalUPS*100)/(_totalUPS+_totalDowns)))
        __evolve["downs"].append(((_totalDowns*100)/(_totalUPS+_totalDowns)))
        __evolve["time"].append(t)
        __evolveCount += 1
        __down0up1.append(0)
    else:
        _downs += 1
        _totalDowns += 1
        _additiveDowns += 1
        __evolve["ups"].append(((_totalUPS*100)/(_totalUPS+_totalDowns)))
        __evolve["downs"].append(((_totalDowns*100)/(_totalUPS+_totalDowns)))
        __evolve["time"].append(t)
        __evolveCount += 1
        __down0up1.append(1)

    #print("shot ("+str(shot)+"): " + str(((_additiveUps*100)/(_additiveUps+_additiveDowns)))[0:6] + "|Up>" + " + " + str(((_additiveDowns*100)/(_additiveUps+_additiveDowns)))[0:6] + "|Down>")
    #print("shot ("+str(shot)+"): " + str(_ups) + "|Up>" + " + " + str(_downs) + "|Down>")


print(str(((_totalUPS*100)/(_totalUPS+_totalDowns))) + "|UPs>" + str(((_totalDowns*100)/(_totalUPS+_totalDowns))) + "|Downs>")
#print("total: " + str(_totalUPS) + "|Ups>" + " + " + str(_totalDowns) + "|Downs>")
```

```python
_additiveUps2 = 0
_additiveDowns2 = 0
shotsCount2 = 50000
for shot2 in range(0, shotsCount2):
    _ups2 = 0
    _downs2 = 0
    t2 = random.uniform(0, 9999999)
    x2 = A2 * math.sin(OMEGA2 * t2 + Phi2)
    if x2 > 0:
        _ups2 += 1
        _totalUPS2 += 1
        _additiveUps2 += 1
        __evolve2["ups"].append(((_totalUPS2*100)/(_totalUPS2+_totalDowns2)))
        __evolve2["downs"].append(((_totalDowns2*100)/(_totalUPS2+_totalDowns2)))
        __evolve2["time"].append(t2)
        __evolveCount2 += 1
        __2down0up1.append(0)
    else:
        _downs2 += 1
        _totalDowns2 += 1
        _additiveDowns2 += 1
        __evolve2["ups"].append(((_totalUPS2*100)/(_totalUPS2+_totalDowns2)))
        __evolve2["downs"].append(((_totalDowns2*100)/(_totalUPS2+_totalDowns2)))
        __evolve2["time"].append(t2)
        __evolveCount2 += 1
        __2down0up1.append(1)

    #print("shot ("+str(shot)+"): " + str(((_additiveUps*100)/(_additiveUps+_additiveDowns)))[0:6] + "|Up>" + " + " + str(((_additiveDowns*100)/(_additiveUps+_additiveDowns)))[0:6] + "|Down>")
    #print("shot ("+str(shot)+"): " + str(_ups) + "|Up>" + " + " + str(_downs) + "|Down>")


print(str(((_totalUPS2*100)/(_totalUPS2+_totalDowns2))) + "|UPs>" + str(((_totalDowns2*100)/(_totalUPS2+_totalDowns2))) + "|Downs>")
#print("total: " + str(_totalUPS) + "|Ups>" + " + " + str(_totalDowns) + "|Downs>")
```

```python
_uu = 0
_dd = 0
_ud = 0
_du = 0
for no in range(0, min(len(__down0up1), len(__2down0up1))):
    if __down0up1[no] == 1 and __2down0up1[no] == 1:
        _uu += 1
    elif __down0up1[no] == 0 and __2down0up1[no] == 0:
        _dd += 1
    elif __down0up1[no] == 1 and __2down0up1[no] == 0:
        _ud += 1
    elif __down0up1[no] == 0 and __2down0up1[no] == 1:
        _du += 1

print("total % : " + str(((_uu+_dd+_ud+_du)*100)/(_uu+_dd+_ud+_du)))
print(str((_uu*100)/(_uu+_dd+_ud+_du)) + "|11> + " + str((_dd*100)/(_uu+_dd+_ud+_du)) + "|00> + " + str((_ud*100)/(_uu+_dd+_ud+_du)) + "|10> + " + str((_du*100)/(_uu+_dd+_ud+_du)) + "|01>")
```
